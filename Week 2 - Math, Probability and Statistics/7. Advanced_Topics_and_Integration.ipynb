1. Topics Overview
1.1 Advanced Statistics
ANOVA (Analysis of Variance):

Compares the means of three or more groups to determine if they are significantly different.

Hypotheses:

Null (H0): All group means are equal.

Alternative (H1): At least one group mean is different.

One-Way ANOVA: Compares one independent variable across groups.

Two-Way ANOVA: Considers two independent variables and their interaction.

Non-Parametric Methods:


Statistical methods that do not assume a specific data distribution.

Examples:

Mann-Whitney U Test: Compares two groups without assuming normality.

Kruskal-Wallis Test: Non-parametric equivalent of one-way ANOVA.

Decision Trees:


Supervised machine learning model.

Splits data into branches based on decision rules to predict outcomes.

Advantage: Easy to interpret and visualize.

ANOVA

Scenario: Analyze if three groups of students (Group A, B, and C) have significantly different exam scores.

```python
from scipy.stats import f_oneway

# Sample data: Exam scores for 3 groups
group_a = [85, 86, 88, 90, 85]
group_b = [78, 80, 82, 79, 77]
group_c = [92, 94, 96, 91, 89]

# One-Way ANOVA
f_stat, p_value = f_oneway(group_a, group_b, group_c)
print("F-Statistic:", f_stat)
print("P-Value:", p_value)

# Decision
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

Explanation:

F-Statistic: Ratio of between-group variance to within-group variance.

P-Value: Probability of observing the result if H0is true.

Decision Tree for Classification

Scenario: Predict if a student will pass (1) or fail (0) based on study hours and attendance.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

# Dataset
data = {
    "Study Hours": [2, 3, 4, 5, 6, 7, 8, 9],
    "Attendance": [1, 0, 1, 1, 1, 0, 1, 1],
    "Pass": [0, 0, 1, 1, 1, 0, 1, 1]
}
df = pd.DataFrame(data)

# Features and target
X = df[["Study Hours", "Attendance"]]
y = df["Pass"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train decision tree model
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Visualize the tree
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=["Study Hours", "Attendance"], class_names=["Fail", "Pass"], filled=True)
plt.show()
```


Explanation:


Decision Tree:

Splits data based on rules (e.g., Study Hours > 4).

Predicts outcomes (pass or fail).

Accuracy: Measures the proportion of correct predictions.

Non-Parametric Test: Kruskal-Wallis

Scenario: Compare exam scores across three groups without assuming normality.

```python
from scipy.stats import kruskal

# Exam scores for 3 groups
group_a = [85, 86, 88, 90, 85]
group_b = [78, 80, 82, 79, 77]
group_c = [92, 94, 96, 91, 89]

# Kruskal-Wallis Test
h_stat, p_value = kruskal(group_a, group_b, group_c)
print("H-Statistic:", h_stat)
print("P-Value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group median is different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

Explanation:

Non-parametric equivalent of ANOVA.

Compares medians rather than means.

Example of AI Applications

Clustering: K-Means

Goal: Group data points into clusters based on similarity.

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
data = np.random.rand(100, 2)

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)

# Cluster assignments and centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Plot
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.scatter(centers[:, 0], centers[:, 1], color='red', marker='X', s=200, label='Centers')
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```
